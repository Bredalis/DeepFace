```python
"""Extract and display detected faces using DeepFace."""

from deepface import DeepFace
import matplotlib.pyplot as plt
```

```python
# Image paths
IMG_PATH_1 = "/content/juhoon.jpg"
IMG_PATH_2 = "/content/sunoo.jpg"
```

```python
def extract_faces(img_path):
    """Extract faces from an image using DeepFace."""
    faces = DeepFace.extract_faces(img_path=img_path)
    return faces
```

```python
def show_faces(face1, face2):
    """Display two faces side by side."""
    plt.figure(figsize=(6, 3))

    plt.subplot(1, 2, 1)
    plt.imshow(face1)
    plt.axis("off")
    plt.title("Face 1")

    plt.subplot(1, 2, 2)
    plt.imshow(face2)
    plt.axis("off")
    plt.title("Face 2")

    plt.show()
```

```python
def main():
    """Main execution function."""
    faces1 = extract_faces(IMG_PATH_1)
    faces2 = extract_faces(IMG_PATH_2)

    face1 = faces1[0]["face"]
    face2 = faces2[0]["face"]

    show_faces(face1, face2)
```

```python
if __name__ == "__main__":
    main()
```
