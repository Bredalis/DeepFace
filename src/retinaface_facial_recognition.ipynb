```python
"""
Display an image using OpenCV and Matplotlib.
"""

from retinaface import RetinaFace
import cv2
import matplotlib.pyplot as plt
```

```python
# Image path
IMG_PATH = "/content/ahyeon.jpg"
```

```python
def detect_faces(image_path):
  result = RetinaFace.detect_faces(image_path)
  print(result.keys())
```

```python
def show_image(image_path):
    """Load and display an image."""
    img = cv2.imread(image_path)

    if img is None:
        print("Error: Could not load image.")
        return

    # Convert BGR to RGB
    img_rgb = img[:, :, ::-1]

    plt.figure(figsize=(10, 10))
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.show()
```

```python
def main():
    """Main function."""
    detect_faces(IMG_PATH)
    show_image(IMG_PATH)
```

```python
if __name__ == "__main__":
    main()
```
