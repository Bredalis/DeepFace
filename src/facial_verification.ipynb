```python
"""
Verify whether two images belong to the same person using DeepFace.
"""

from deepface import DeepFace
```

```python
# Image paths
IMG1_PATH = "/content/ahyeon.jpg"
IMG2_PATH = "/content/sunoo.jpg"
```

```python
def verify_faces(img1_path, img2_path):
    """Verify if two faces belong to the same person."""
    result = DeepFace.verify(
        img1_path=img1_path,
        img2_path=img2_path,
        model_name="ArcFace",
        enforce_detection=False
    )
    return result
```

```python
def main():
    """Main function."""
    result = verify_faces(IMG1_PATH, IMG2_PATH)
    print(result)
```

```python
if __name__ == "__main__":
    main()
```
