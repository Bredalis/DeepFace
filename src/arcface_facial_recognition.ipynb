```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
# Image paths
img_path1 = "/content/wonyoung.jpg"
img_path2 = "/content/hanni.jpg"
```

```python
# Verify if the faces belong to the same person
result = DeepFace.verify(
    img_path1,
    img_path2,
    model_name="ArcFace",
    detector_backend="retinaface"
)
```

```python
result.keys()
```

```python
# Load images
img1 = cv2.imread(img_path1)
img2 = cv2.imread(img_path2)
```

```python
# Display images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(img1[:, :, ::-1])
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img2[:, :, ::-1])
plt.axis("off")

plt.suptitle(f"Are they the same face?: {result['verified']}")
plt.show()
```
