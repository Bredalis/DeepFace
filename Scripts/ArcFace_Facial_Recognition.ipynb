```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img_path1 = "/content/Wonyoung.jpg"
img_path2 = "/content/Wonyoung2.jpg"
```

```python
# Verificar si las caras son las mismas
obj = DeepFace.verify(
    img_path1, img_path2,
    model_name = "ArcFace",
    detector_backend = "retinaface"
)
```

```python
obj.keys()
```

```python
img1 = cv2.imread(img_path1)
img2 = cv2.imread(img_path2)
```

```python
plt.figure(figsize = (8, 4))

plt.subplot(1, 2, 1)
plt.imshow(img1[:, :, ::-1])
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img2[:, :, ::-1])
plt.axis("off")

plt.suptitle(f"Â¿Son la misma cara?: {obj['verified']}")
plt.show()
```
